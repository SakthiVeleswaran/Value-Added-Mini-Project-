## Normalization

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
import numpy as np
```

```python
normalization_df = pd.DataFrame({'X1':np.random.randint(-100,100,1000),
                                 'X2':np.random.randint(-80,80,1000),
                                 'X3':np.random.randint(-150,150,1000)})
```

```python
normalization_df
```

```python
ax = plt.axes(projection = '3d')
ax.scatter3D(normalization_df.X1, normalization_df.X2, normalization_df.X3)
plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normal = Normalizer()
```

```python
normalization_df = normal.fit_transform(normalization_df)
```

```python
normalization_df
```

```python
norm_df = pd.DataFrame(normalization_df,columns=['X1','X2','X3'])        
```

```python
norm_df
```

```python
ax = plt.axes(projection = '3d')
ax.scatter3D(norm_df.X1, norm_df.X2, norm_df.X3)
plt.figure()
```

```python
data = {'First Score':[100,90,np.nan,95],
        'Second Score':[30,45,56,np.nan],
        'Third Score':[np.nan,40,80,98]}
```

```python
scores
```

```python
scores.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```
