## Creating data

```python
movies = ['JL','Batman','Avengers','Ironman','Mark']
percentage = [23,31,30,19,7]
```

## Importing Libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Scatter Plot

```python
plt.scatter(movies, percentage)
```

## Line plot

```python
plt.plot(movies, percentage, linestyle='dashed')
```

```python
plt.plot(movies, percentage, linestyle='solid')
```

```python
plt.plot(movies, percentage, linestyle='dotted')
```

```python
plt.plot(movies, percentage, linestyle='dashdot')
```

```python
plt.pie(percentage, labels = movies)
```

## Bar Chart

```python
plt.bar(movies, percentage)
```
